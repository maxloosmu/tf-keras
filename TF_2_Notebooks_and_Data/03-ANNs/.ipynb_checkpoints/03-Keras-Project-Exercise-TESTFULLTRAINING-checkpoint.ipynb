```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# might be needed depending on your version of Jupyter
%matplotlib inline
df = pd.read_csv('../DATA/lending_club_loan_two.csv')
sns.countplot(data=df, x="loan_status")
plt.xlabel("Loan Status")
plt.ylabel("Count")
plt.title("Countplot of Loan Status")
plt.show()
```

```python
sns.histplot(data=df, x="loan_amnt", bins=50, kde=True)
plt.xlabel("Loan Amount")
plt.ylabel("Frequency")
plt.title("Histogram of Loan Amount")
plt.show()
```

```python
unique_grades = df['grade'].unique()
sorted_unique_grades = sorted(unique_grades)
print("Unique values in the 'grade' column:", sorted_unique_grades)

# For 'sub_grade' column
unique_subgrades = df['sub_grade'].unique()
sorted_unique_subgrades = sorted(unique_subgrades)
print("Unique values in the 'sub_grade' column:", sorted_unique_subgrades)
```

```python
def map_loan_status(status):
    if status == "Fully Paid":
        return 1
    elif status == "Charged Off":
        return 0
    else:
        return None  # Optional: Handle other statuses
df['loan_repaid'] = df['loan_status'].apply(map_loan_status)
df[['loan_repaid', 'loan_status']]
```

```python
# Using .isna() or .isnull() to create a boolean DataFrame
missing_values = df.isna()  # or df.isnull()

# Using .sum() to count missing values per column
missing_values_count = missing_values.sum()

# Display the Series
print(missing_values_count)
```

```python
df.drop('emp_title', axis=1, inplace=True)
```

```python
df.drop('emp_length', axis=1, inplace=True)
df.drop('title', axis=1, inplace=True)
# Create a Series with mean 'mort_acc' values per 'total_acc' value
mort_acc_mean_by_total_acc = df.groupby('total_acc')['mort_acc'].mean()

# Define a function to fill missing 'mort_acc' values
def fill_missing_mort_acc(row):
    if pd.isna(row['mort_acc']):
        return mort_acc_mean_by_total_acc[row['total_acc']]
    else:
        return row['mort_acc']

# Apply the function to fill missing 'mort_acc' values
df['mort_acc'] = df.apply(fill_missing_mort_acc, axis=1)
```

```python
missing_values = df.isna()
missing_values_count = missing_values.sum()
print(missing_values_count)
```

```python
df.dropna(subset=['revol_util', 'pub_rec_bankruptcies'], inplace=True)
df.isnull().sum()
```

```python
# Convert 'sub_grade' into dummy variables with drop_first=True
sub_grade_dummies = pd.get_dummies(df['sub_grade'], drop_first=True)

# Concatenate the new columns to the original DataFrame
df = pd.concat([df, sub_grade_dummies], axis=1)

# Drop the original 'sub_grade' column
df.drop('sub_grade', axis=1, inplace=True)
```

```python
df.drop('grade', axis=1, inplace=True)
```

```python
# List of columns to convert into dummy variables
columns_to_convert = ['verification_status', 'application_type', 'initial_list_status', 'purpose']

# Convert the columns into dummy variables with drop_first=True
dummies = pd.get_dummies(df[columns_to_convert], drop_first=True)

# Concatenate the new columns to the original DataFrame
df = pd.concat([df, dummies], axis=1)

# Drop the original columns
df.drop(columns=columns_to_convert, axis=1, inplace=True)
df.columns
```

```python
# Convert 'home_ownership' into dummy variables with custom mapping and drop_first=True
df['home_ownership'] = df['home_ownership'].replace(['NONE', 'ANY'], 'OTHER')
home_ownership_dummies = pd.get_dummies(df['home_ownership'], drop_first=True)

# Concatenate the new columns to the original DataFrame
df = pd.concat([df, home_ownership_dummies], axis=1)

# Drop the original 'home_ownership' column
df.drop('home_ownership', axis=1, inplace=True)
print(df.columns)
print(df.select_dtypes(['object']).columns)
```

```python
df['zip_code'] = df['address'].apply(lambda address:address[-5:])
zip_code_dummies = pd.get_dummies(df['zip_code'],drop_first=True)
df = df.drop(['zip_code','address'],axis=1)
df = pd.concat([df, zip_code_dummies],axis=1)
print(df.columns)
print(df.select_dtypes(['object']).columns)
```

```python
df = df.drop('issue_d',axis=1)
# Extract the year from 'earliest_cr_line' and convert to a numeric feature
df['earliest_cr_year'] = df['earliest_cr_line'].apply(lambda x: int(x.split('-')[1]))
# Drop the original 'earliest_cr_line' feature
df.drop('earliest_cr_line', axis=1, inplace=True)
df.select_dtypes(['object']).columns
df = df.drop('loan_status',axis=1)
```

```python
df['term'] = df['term'].apply(lambda term: int(term[:3]))
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.constraints import max_norm
from tensorflow.keras.callbacks import EarlyStopping
model = Sequential()

# input layer
model.add(Dense(78,  activation='relu'))
model.add(Dropout(0.2))

# hidden layer
model.add(Dense(39, activation='relu'))
model.add(Dropout(0.2))

# hidden layer
model.add(Dense(19, activation='relu'))
model.add(Dropout(0.2))

# output layer
model.add(Dense(units=1,activation='sigmoid'))

# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam')
```
